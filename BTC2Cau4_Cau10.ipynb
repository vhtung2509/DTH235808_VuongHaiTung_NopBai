```python
# Ý nghĩa các toán tử trong Python

a = 10
b = 3

# /  : Chia lấy kết quả là số thực
ket_qua_chia = a / b        # 10 / 3 = 3.333...

# // : Chia lấy phần nguyên
ket_qua_chia_nguyen = a // b  # 10 // 3 = 3

# %  : Chia lấy phần dư
phan_du = a % b             # 10 % 3 = 1

# ** : Lũy thừa
luy_thua = a ** b           # 10 ** 3 = 1000

# and: Toán tử logic "và"
ket_qua_and = (a > 5) and (b < 5)  # True and True = True

# or : Toán tử logic "hoặc"
ket_qua_or = (a < 5) or (b < 5)    # False or True = True

# is : So sánh đối tượng (kiểm tra cùng tham chiếu bộ nhớ)
x = [1, 2]
y = x
z = [1, 2]
print(x is y)  # True (cùng đối tượng)
print(x is z)  # False (khác đối tượng, dù giá trị giống nhau)
```

```python
# Khai báo biến
i1 = 2
i2 = 5
i3 = -3
d1 = 2.0
d2 = 5.0
d3 = -0.5

# (a) i1 + (i2 * i3)
# 2 + (5 * -3) = 2 + (-15) = -13
print("(a)", i1 + (i2 * i3))  # -13

# (b) i1 * (i2 + i3)
# 2 * (5 + -3) = 2 * 2 = 4
print("(b)", i1 * (i2 + i3))  # 4

# (c) i1 / (i2 + i3)
# 2 / (5 + -3) = 2 / 2 = 1.0
print("(c)", i1 / (i2 + i3))  # 1.0

# (d) i1 // (i2 + i3)
# 2 // (5 + -3) = 2 // 2 = 1
print("(d)", i1 // (i2 + i3))  # 1

# (e) i1 / i2 + i3
# 2 / 5 + (-3) = 0.4 + (-3) = -2.6
print("(e)", i1 / i2 + i3)  # -2.6

# (f) i1 // i2 + i3
# 2 // 5 + (-3) = 0 + (-3) = -3
print("(f)", i1 // i2 + i3)  # -3

# (g) 3 + 4 + 5 / 3
# 3 + 4 + (5 / 3) = 7 + 1.666... = 8.666...
print("(g)", 3 + 4 + 5 / 3)  # 8.666...

# (h) 3 + 4 + 5 // 3
# 3 + 4 + (5 // 3) = 7 + 1 = 8
print("(h)", 3 + 4 + 5 // 3)  # 8

# (i) (3 + 4 + 5) / 3
# (3 + 4 + 5) = 12; 12 / 3 = 4.0
print("(i)", (3 + 4 + 5) / 3)  # 4.0

# (j) (3 + 4 + 5) // 3
# (3 + 4 + 5) = 12; 12 // 3 = 4
print("(j)", (3 + 4 + 5) // 3)  # 4

# (k) d1 + (d2 * d3)
# 2.0 + (5.0 * -0.5) = 2.0 + (-2.5) = -0.5
print("(k)", d1 + (d2 * d3))  # -0.5

# (l) d1 + d2 * d3
# 2.0 + (5.0 * -0.5) = 2.0 + (-2.5) = -0.5
print("(l)", d1 + d2 * d3)  # -0.5

# (m) d1 / d2 - d3
# (2.0 / 5.0) - (-0.5) = 0.4 + 0.5 = 0.9
print("(m)", d1 / d2 - d3)  # 0.9

# (n) d1 / (d2 - d3)
# 2.0 / (5.0 - (-0.5)) = 2.0 / 5.5 = 0.3636...
print("(n)", d1 / (d2 - d3))  # 0.363636...

# (o) d1 + d2 + d3 / 3
# d3 / 3 = -0.5 / 3 = -0.166...; 2.0 + 5.0 + (-0.166...) = 6.833...
print("(o)", d1 + d2 + d3 / 3)  # 6.833...

# (p) (d1 + d2 + d3) / 3
# (2.0 + 5.0 + -0.5) = 6.5; 6.5 / 3 = 2.166...
print("(p)", (d1 + d2 + d3) / 3)  # 2.166...

# (q) d1 + d2 + (d3 / 3)
# d3 / 3 = -0.166...; 2.0 + 5.0 + (-0.166...) = 6.833...
print("(q)", d1 + d2 + (d3 / 3))  # 6.833...

# (r) 3 * (d1 + d2) * (d1 - d3)
# (d1 + d2) = 7.0; (d1 - d3) = 2.0 - (-0.5) = 2.5; 3 * 7.0 * 2.5 = 52.5
print("(r)", 3 * (d1 + d2) * (d1 - d3))  # 52.5
```
